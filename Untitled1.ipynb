```python
from pyquery import PyQuery as pq
```

```python
with open("data/regierungsprogramm-2013-18-clean.html") as f:
    q=pq(f.read())
```

```python
sections=q("section")
```

```python
def get_titles(s,level):
    t=pq(s)
    return t("h%s"%level)
```

```python
get_titles(sections[4],1)[0].text
```

```python
import re
def slugify(s):
    return re.sub("[^a-z0-9]","-",s.lower())
```

```python
titles=q("h1, h2")
```

```python
for t in titles:
    t.set("id",slugify(t.text))
```

```python
q.html
```

```python
q.html()
```

```python
def htmlify(e):
    return u"<a href='#%s'>%s</a>"%(e.get("id"),e.text)    
```

```python
htmlify(titles[3])
```

```python
def make_titles(sections):
    return "\n".join(["\n".join(["<li>%s<ul>%s</ul></li>"%(htmlify(e),"\n".join(["<li>%s</li>"%htmlify(f) for f in get_titles(s,2)])) for e in get_titles(s,1)]) for s in sections])
```

```python
make_titles(sections)
```

```python
q("div#toc").html(make_titles(sections))
```

```python
q.html()
```
